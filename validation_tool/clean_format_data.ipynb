### Clean 2021 A-10 data  
This is a 1-time cleaning process since all other data will come from Black Cat (This is from NTD tables). Only cleaning enough to get started making the prototype.  
It is in a weird format of a) identifying columns, then b) cols for facility type, then c) cols for the ownership type. BUT there might be > 1 ownership type in which case its unclear unless you look at it row by row, which facilities are under which type of ownership.  
  
We make it wide form, 1 column per ownership type + facility type combo. Ignore those with > 1 ownership type.   
Additional 2 columns for transport type.

```python
import pandas as pd
import numpy as np 
```

```python
# Data found from NTD tables by Slalom team.
df = pd.read_csv("data/a10_2021.csv") # A-10
df = df.drop(['Directly Operated Ownership Types:', 'Purchased Transportation Ownership Types:'], axis=1)
df.head()
```

Count the non-zero values in the columns on ownership type.  
https://stackoverflow.com/questions/69605383/pandas-determine-column-labels-that-contribute-to-non-zero-values-in-each-row

```python
df['own_type_count'] = np.count_nonzero(df.iloc[:, 12:19],axis=1)
df.head()
```

```python
df['own_type_count'].value_counts()
```

```python
df['own_type_label'] = df.iloc[:, 12:19].apply(lambda r: r.index[r.ne(0)].to_list(), axis=1)
df.head()
```

Make a new df with the format we want, called `df_wide`. Exploded out columns for ownership type and facility type, so there's one column per combination.

```python
list(df.columns.values)
ownership_types = list(df.columns.values)[12:19]
print(ownership_types)
fac_types = list(df.columns.values)[7:12]
print(fac_types)
```

```python
colnames = [x + "_" + y for x in ownership_types for y in fac_types]
colnames
```

```python
# First make new empty dataframe - with same index as old dataframe
df_wide = df.iloc[:, 0:7]
df_wide.loc[:, colnames] = 0
df_wide.head()
```

**Logic:** If own_type_count == 1 then add to the new df: cols 1-5 in the cols named `f"{own_type_label}_{fac_type}"`

```python
### here we've already made df_wide
# Subset df by those with 1 ownership listed
one_ownership = df[df['own_type_count']==1]
print(one_ownership.shape)
one_ownership.head()
```

```python
# Check whether there are duplicates in the index, given the error in my function below.
one_ownership.index.has_duplicates
```

Inconsistent logic around rows. For each row:    
* if only 1 value in `df.iloc[x, 10:17]` (row x, cols 10:17 by index) is not null, then put the values in `df.iloc[x, 5:10]` in the columns matching the name `f"{ownership_type}_{fac_type}"`

```python
# Make a df 1 row wide at a time
# put in a list.         frames = [df1, df2, df3]
# concat all at the end. result = pd.concat(frames)

from itertools import islice

df_cleaned = []
for index,row in islice(one_ownership.iterrows(), 0, 273, 1): #273
#     print(index)
    ownership_type = row['own_type_label'][0] + "_"
    responses = one_ownership.iloc[index,7:12]
    responses = responses.add_prefix(ownership_type) #now a series of form responses
    
    row_agency_info = one_ownership.iloc[index, 0:7]
    row_form_info = pd.Series(index=colnames, dtype='float').fillna(0)
    ### now replace appropriate cols with values in responses
    cols_to_fill = [x for x in colnames if x.startswith(ownership_type)]
    row_form_info.loc[cols_to_fill] = responses
    complete_row_info = pd.concat([row_agency_info, row_form_info], names = colnames, axis = 0)
#     print(complete_row_info)
#     print(f"complete_row_info is {type(complete_row_info)}")
    df_cleaned.append(complete_row_info)
    
final_data = pd.DataFrame(df_cleaned)
# print(type(df_cleaned))
# print(df_cleaned[0].colnames)
print(final_data.shape)
print(type(final_data))
```
